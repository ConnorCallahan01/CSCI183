## 3. Find the correlation matrix for this dataset. Report which features tend to have a high correlation with the target variable.

### For the housing.csv data: 

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

housingCorr = housing.corr()
#visualize the data to see where the correlations exist
plt.subplots(figsize=(20,20))
sns.heatmap(
    housingCorr,
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(22, 220, n=150),
    annot=True, annot_kws={'size': 15},
    square=True
)

#initiate the array and dictionary to be used for correlation analysis: 
temp_array = {}
temp_list = []
for line in housingCorr: 
    temp_array[line] = []


# Make an array of r-values that represent high correlation and dictionary of those
def rArray(t_list, t_array, corr):
    for i in corr: #go through correlation matrix by column
        for j in corr: #go through correlation matrix by row
            value = corr[i][j] #set the cell value to value
            if  value > .99999: #if the value is 1 or greater
                break
            elif value > 0.65 or value < -0.65: #if the value would represent a correlation
                t_list.append(value) #put the value in the array of r-values
                t_array[j].append(i) #note the correlation in the dictionary        
    return t_list
    return t_array
#     print()         
#     print(t_list)
#     print()

#
#Print the correlation observations: 
def corrObserve(t_array, t_list):
    n = 0
    for i in t_array: #access the dictionary
        if t_array[i] != []: #if the dictionary is not empty
            for j in t_array[i]: #go through each value per key
                if t_list[n] < 1 and t_list[n] > 0: #if the corresponding value in the correlation array is positive:
                    print("\u0332".join(i), " has a strong, positive correlation to", "\u0332".join(j), "with an r value of: ", t_list[n]) #print the relation and the r value
                    print()
                elif t_list[n] > -1 and t_list[n] < 0: #if the corresponding value in the correlation array is negative:
                    print(i," has a strong, negative correlation to ", j, " with an r value of: ", t_list[n]) #print the relation and the r value
                    print()
                n += 1

rArray(temp_list, temp_array, housingCorr)
corrObserve(temp_array, temp_list)    
        
    
```

### For the kc_house_data.csv data: 

```python
kcHousingCorr = kcHousing.corr() 

#visualize the data to see where the correlations exist
plt.subplots(figsize=(20,20))
sns.heatmap(
    kcHousingCorr,
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(100, 330, n=150),
    annot=True, annot_kws={'size': 10},
    square=True
)

#initiate the array and dictionary to be used for correlation analysis: 
temp_array = {}
temp_list = []
for line in kcHousingCorr: 
    temp_array[line] = []


rArray(temp_list, temp_array, kcHousingCorr)
corrObserve(temp_array, temp_list) 
```

## 6. Selecting different features from step 5, implement a linear regression algorithm and find the slope, the intercept and the error of the regression model.

## For Housing

```python
import numpy as np
from scipy.stats import linregress

temp_array = {}
temp_list = []
for line in housingCorr: 
    temp_array[line] = []

rArray(temp_list, temp_array, housingCorr)
def findSIError(t_array):
    for i in t_array:
        X = housing[i]
        for j in t_array[i]:
                if len(t_array[i]) != 0:
                    Y = housing[j]
                    mask = ~np.isnan(X) & ~np.isnan(Y)
                    result = linregress(X[mask], Y[mask])
                    print("Slope = ",result.slope, "; Intercept = ", result.intercept, "; Error = ", result.stderr)
                    plt.plot(X, Y, 'o', label='original data')
                    plt.plot(X, result.intercept + result.slope*X, 'r', label='fitted line')
                    plt.legend()
                    plt.xlabel(i)
                    plt.ylabel(j)
                    plt.show()
                    
findSIError(temp_array)
```

## 9. Create a table similar to one given below for all the features selected in step 6

## For Housing

```python
import numpy as np
from scipy.stats import linregress
import pandas as pd

temp_array = {}
temp_list = []
for line in housingCorr: 
    temp_array[line] = []

house_regress = {"Feature": [], "Slope": [], "Intercept": [], "Error": []}
rArray(temp_list, temp_array, housingCorr)
def findSIError(t_array):
    for i in t_array:
        X = housing[i]
        for j in t_array[i]:
                if len(t_array[i]) != 0:
                    Y = housing[j]
                    mask = ~np.isnan(X) & ~np.isnan(Y)
                    result = linregress(X[mask], Y[mask])
                    house_regress["Feature"].append(j + ' and ' + i)
                    house_regress["Slope"].append(result.slope)
                    house_regress["Intercept"].append(result.intercept)
                    house_regress["Error"].append(result.stderr)
                    #print("Slope = ",result.slope, "; Intercept = ", result.intercept, "; Error = ", result.stderr)
        
    house_regress_df = pd.DataFrame.from_dict(house_regress)
    return house_regress_df

findSIError(temp_array)
```
